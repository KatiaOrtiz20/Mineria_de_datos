**INFORMACIÓN ESTADÍSTICA BÁSICA**

```python
print("La suma del income es: ",dc['Income'].sum())
print("\n")
print("El income promedio es: ",dc['Income'].mean())
print("\n")
print("La suma acumulada de la columna fila por fila de MntFruits es:")
print(dc['MntFruits'].cumsum())
print("\n")
print("Este es el summary de la columna Income:")
print(dc['Income'].describe())
print("\n")
print("Estos son los elementos no nulos en la columna Recency: ")
print(dc['Recency'].count())
print("\n")
print("El año de nacimineto minimo es:",dc['Year_Birth'].min())
print("El año de nacimiento maxim es:",dc['Year_Birth'].max())
print("\n")
print("El año mediano es",dc['Year_Birth'].median())
print("La varianza del Income es",dc['Income'].var())
print("La desviación estandar del Income es",dc['Income'].std())
print("\n")
print("El valor de simetria del Income es",dc['Income'].skew())
print("\n")
print("Las caracteristicas de distribucion de frecuencias del Income es:",dc['Income'].kurt())
print("\n")
print("Correlacion de los datos:\n")
dc.corr()
```

**INFORMACIÓN ESTADÍSTICA BÁSICA**

```python
print("La suma de los serum sodium es: ",df['serum_sodium'].sum())
print("\n")
print("La edad promedio es: ",df['age'].mean())
print("\n")
print("La suma acumulada de la columna fila por fila de serum sodium es:")
print(df['serum_sodium'].cumsum())
print("\n")
print("Este es el summary de la columna Age:")
print(df['age'].describe())
print("\n")
print("Estos son los elementos no nulos en la columna anemia: ",df['anaemia'].count())
print("\n")
print("La edad minima es:",df['age'].min())
print("La edad maxima es:",df['age'].max())
print("\n")
print("La edad mediana es",df['age'].median())
print("La varianza de la edad es",df['age'].var())
print("La desviación estandar de la edad es",df['age'].std())
print("\n")
print("El valor de simetria en la edad es",df['age'].skew())
print("\n")
print("Las caracteristicas de distribucion de frecuencias de la edad es:",df['age'].kurt())
print("\n")
print("Correlacion de los datos:\n")
df.corr()
```
